```python
import pandas as pd
df = pd.read_csv('/content/Cost_of_Living_Index_by_Country_2024.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Country'], df['Cost of Living Index'], marker='o', label='Cost of Living Index')
plt.plot(df['Country'], df['Rent Index'], marker='s', label='Rent Index')
plt.xlabel('Country')
plt.ylabel('Index Value')
plt.title('Cost of Living Index and Rent Index with Markers')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
df_head = df.head(5)
plt.figure(figsize=(12, 6))
plt.plot(df_head['Country'], df_head['Cost of Living Index'], marker='o', label='Cost of Living Index')
plt.plot(df_head['Country'], df_head['Rent Index'], marker='s', label='Rent Index')
plt.xlabel('Country')
plt.ylabel('Index Value')
plt.title('Cost of Living Index and Rent Index (First 5 Rows)')
plt.legend()
plt.show()
```
